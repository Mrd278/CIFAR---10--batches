```
# Importing Libraries

import pandas as pd
import numpy as np
from keras.layers import Dense,Dropout,Flatten
from keras.layers import Conv2D,MaxPooling2D
from keras.models import Sequential
from keras.utils import np_utils
```

```
# Function to convert the data into dictionary

def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```
ls
```

```
# Calling unpickle function 

train_1 = unpickle('data_batch_1')
train_2 = unpickle('data_batch_2')
train_3 = unpickle('data_batch_3')
train_4 = unpickle('data_batch_4')
train_5 = unpickle('data_batch_5')
test = unpickle('test_batch')
```

```
'''Extraction of data column from dictionary
   And converting as numpy array
''' 

data_1 = train_1[b'data']
data_1 = np.asarray(data_1)
data_2 = train_2[b'data']
data_2 = np.asarray(data_2)
data_3 = train_3[b'data']
data_3 = np.asarray(data_3)
data_4 = train_4[b'data']
data_4 = np.asarray(data_4)
data_5 = train_5[b'data']
data_5 = np.asarray(data_5)
test_data = test[b'data']
test_data = np.asarray(test_data)
```

```
# Collecting all the image Data into image_train_data and image_test_data

image_train_data = np.zeros((50000,1024,3))

image_train_data[:10000,:,0] = data_1[:,:1024]
image_train_data[:10000,:,1] = data_1[:,1024:2048]
image_train_data[:10000,:,2] = data_1[:,2048:3072]

image_train_data[10000:20000,:,0] = data_2[:,:1024]
image_train_data[10000:20000,:,1] = data_2[:,1024:2048]
image_train_data[10000:20000,:,2] = data_2[:,2048:3072]

image_train_data[20000:30000,:,0] = data_3[:,:1024]
image_train_data[20000:30000,:,1] = data_3[:,1024:2048]
image_train_data[20000:30000,:,2] = data_3[:,2048:3072]

image_train_data[30000:40000,:,0] = data_4[:,:1024]
image_train_data[30000:40000,:,1] = data_4[:,1024:2048]
image_train_data[30000:40000,:,2] = data_4[:,2048:3072]

image_train_data[40000:50000,:,0] = data_5[:,:1024]
image_train_data[40000:50000,:,1] = data_5[:,1024:2048]
image_train_data[40000:50000,:,2] = data_5[:,2048:3072]

image_test_data = np.zeros((10000,1024,3))

image_test_data[:,:,0] =  test_data[:,:1024]
image_test_data[:,:,1] =  test_data[:,1024:2048]
image_test_data[:,:,2] =  test_data[:,2048:3072]
```

```
# Extraction of labels from the dictionary

label_1 = train_1[b'labels']
label_1 = np.asarray(label_1)

label_2 = train_2[b'labels']
label_2 = np.asarray(label_2)

label_3 = train_3[b'labels']
label_3 = np.asarray(label_3)

label_4 = train_4[b'labels']
label_4 = np.asarray(label_4)

label_5 = train_5[b'labels']
label_5 = np.asarray(label_5)

test_label = test[b'labels']
test_label = np.asarray(test_label)
```

```
# Collecting all the labels into train_label

train_label = np.zeros(50000)
train_label[:10000] = label_1[:]
train_label[10000:20000] = label_2[:]
train_label[20000:30000] = label_3[:]
train_label[30000:40000] = label_4[:]
train_label[40000:50000] = label_5[:]
```

```
# Normalizing the image data

image_train_data = image_train_data/255
image_test_data = image_test_data/255

# Converting the datatype of labels into uint8

test_label = test_label.astype('uint8')
train_label = train_label.astype('uint8')

# Coverting the labels encoded form

test_label = np_utils.to_categorical(test_label,10)
train_label = np_utils.to_categorical(train_label,10)
```

```
# Reshaping the Image data

image_train_data = image_train_data.reshape(50000,32,32,3)
image_test_data = image_test_data.reshape(10000,32,32,3)
```

```
# Building the CNN Model

model = Sequential()
model.add(Conv2D(filters = 16, kernel_size = (5,5), activation = 'relu',input_shape = (32,32,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Conv2D(filters = 32, kernel_size = (5,5), activation = 'relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Flatten())

model.add(Dense(128,activation = 'relu'))
model.add(Dropout(0.25))

model.add(Dense(64,activation = 'relu'))
model.add(Dropout(0.25))

model.add(Dense(32,activation = 'relu'))
model.add(Dropout(0.25))

model.add(Dense(10,activation = 'sigmoid'))
```

```
model.summary()
```

```
# Compiling the Model

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```
# fitting the Data

model.fit(image_train_data,train_label, epochs = 10, validation_data = (image_test_data,test_label), batch_size = 64)
```

```
# Saving the model

model_name = "CIFAR.h5"
model.save(model_name,overwrite = True)
```
